#Logistic Regression method:

```python
data = pd.read_csv("/content/drive/MyDrive/DS102 TH/Lab3/data.csv", sep=";")
data.head()
```

```python
data.isna().sum(axis="index")
```

```python
data = data.dropna()
```

```python
data.duplicated().any()
```

```python
data.info()
```

```python
def LabelEncoder (val:str) -> int:
  if val=="Enrolled" or val=="Dropout":
    return 0

  return 1
```

```python
data["Target"] = data["Target"].apply(LabelEncoder)
data.head()
```

```python
X = data.drop('Target', axis=1)
y = data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
logistic_reg = LogisticRegression(solver='lbfgs', max_iter=100)
logistic_reg.fit(X_train, y_train)
```

```python
y_prob = logistic_reg.predict_proba(X_test)
y_prob
```

```python
y_pred = logistic_reg.predict(X_test)
```

```python
loss = log_loss(y_test, y_prob)
print('Value of the loss function:', loss)
```

```python
acc = accuracy_score(y_test, y_pred)
print('Accuracy of Logistic Regression:', acc)
```

#Softmax Regression method:

```python
data = pd.read_csv("/content/drive/MyDrive/DS102 TH/Lab3/data.csv", sep=";")
data.head()
```

```python
data.isna().sum(axis="index")
```

```python
data = data.dropna()
```

```python
data.duplicated().any()
```

```python
data.info()
```

```python
def LabelEncoder (val:str) -> int:
  if val=="Enrolled":
    return 0
  elif val=="Dropout":
    return 1
  return 2
```

```python
data["Target"] = data["Target"].apply(LabelEncoder)
data.head()
```

```python
X = data.drop('Target', axis=1)
y = data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
sm_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=100)
sm_reg.fit(X_train, y_train)
```

```python
y_proba = sm_reg.predict_proba(X_test)
y_proba
```

```python
y_pred = sm_reg.predict(X_test)
```

```python
loss = log_loss(y_test, y_proba)
print('Value of the loss function:', loss)
```

```python
acc = accuracy_score(y_test, y_pred)
print('Accuracy of Softmax Regression:', acc)
```
